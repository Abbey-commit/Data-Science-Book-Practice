# Analyze and Visualize the Data using Python:

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

```python
# Create a connection to the SQLite database
conn = sqlite3.connect('epl_data.db')
```

```python
# Load the data into DataFrames
top_home_teams = pd.read_sql("SELECT HomeTeam, SUM(FTHG) AS TotalHomeGoals FROM epl GROUP BY HomeTeam ORDER BY TotalHomeGoals DESC LIMIT 5", conn)
top_away_teams = pd.read_sql('SELECT AwayTeam, SUM(FTAG) AS TotalAwayGoals FROM epl GROUP BY AwayTeam ORDER BY TotalAwayGoals DESC LIMIT 5', conn)
top_referees = pd.read_sql('SELECT Referee, COUNT(*) AS MatchesOfficiated FROM epl GROUP BY  Referee ORDER BY MatchesOfficiated DESC LIMIT 5', conn)
```

```python
# Close the connection
conn.close()
```

```python
# Set style
sns.set_style('whitegrid')
```

```python
# Plot top 5 teams with the most goals scored at home
plt.figure(figsize=(10, 5))
sns.barplot(x='TotalHomeGoals', y='HomeTeam', data=top_home_teams, palette='viridis')
plt.title('Top 5 Teams with the Most Goals Scored at Home')
plt.xlabel('Total Home Goals')
plt.ylabel('Home Team')
plt.show();
```

```python
# Plot top 5 teams with the most goals scored away
plt.figure(figsize=(10, 5))
sns.barplot(x='TotalAwayGoals', y='AwayTeam', data=top_away_teams, palette='magma')
plt.title('Top 5 Teams with the Most Goals Scored Away')
plt.xlabel('Total Away Goals')
plt.ylabel('Away Team')
plt.show()
```

```python
# Plot top 5 referees who officiated the most matches
plt.figure(figsize=(10, 5))
sns.barplot(x='MatchesOfficiated', y='Referee', data=top_referees, palette='coolwarm')
plt.title('Top 5 Referees Who Officiated the Most Matches')
plt.xlabel('Matches Officiated')
plt.ylabel('Referees')
plt.show()
```
